```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gammaln
```

```python
# Given parameters
a0, b0 = 1, 0.1  # Prior parameters
a_post, b_post = 43, 5.1  # Posterior parameters

# Lambda range. Avoiding 0 to prevent log(0)
lambda_vals = np.linspace(0.001, 30, 500)

# Compute log densities using gammaln
log_prior = (a0 - 1) * np.log(lambda_vals) - b0 * lambda_vals - gammaln(a0) + a0 * np.log(b0)
log_posterior = (a_post - 1) * np.log(lambda_vals) - b_post * lambda_vals - gammaln(a_post) + a_post * np.log(b_post)

# Convert from log-density to density
prior_pdf = np.exp(log_prior)
posterior_pdf = np.exp(log_posterior)

# Plotting
plt.figure(figsize=(8, 5))
plt.style.use('ggplot')  # Gray background
plt.plot(lambda_vals, prior_pdf, label=r'Prior $p(\lambda)$', linestyle='--', color='brown')
plt.plot(lambda_vals, posterior_pdf, label=r'Posterior $p(\lambda|\mathbf{y})$', color='red')
plt.xlabel(r'$\lambda$')
plt.ylabel('Probability density')
plt.legend()
plt.grid(color='white')  # White gridlines
plt.show()
```
